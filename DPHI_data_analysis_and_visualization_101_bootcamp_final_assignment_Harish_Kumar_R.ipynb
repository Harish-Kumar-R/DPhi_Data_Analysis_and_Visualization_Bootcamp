```
import plotly.express as px
```

```
data_19["Company type"].value_counts()
```

```
pie_chart1 = px.pie(data_19, values=data_19.index, names='Company type',title='Company Types in the year 2019')
pie_chart1.show()
```

Consulting / Agency companies are **not more popular than** Startups

```
salarydata_2018_1 = data_18[{'Years of experience','Current Salary'}]
salarydata_2018_2 = salarydata_2018_1.apply(lambda x: x.fillna(x.value_counts().index[0]))
salarydata_2018_3 = salarydata_2018_1.groupby(['Years of experience']).median()
salarydata_2018_3
```

```
lineplot = px.line(salarydata_2018_3, y="Current Salary", title='Total years of Experience vs Current Salary')
lineplot.show()
```
